```python
import nltk
from nltk.util import bigrams
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import PyPDF2
import matplotlib.pyplot as plt
import numpy as np
nltk.download('punkt')
nltk.download('stopwords')

# Read the PDF file containing papers
pdf_file_path = r'C:\Users\user\Desktop\499B\BERT.pdf'

# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    text = ''
    with open(pdf_path, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        for page in pdf_reader.pages:
            text += page.extract_text()
    return text

# Extract text from PDF
papers_text = extract_text_from_pdf(pdf_file_path)

# Preprocess and clean the text
cleaned_text = ' '.join(word for word in word_tokenize(papers_text.lower()) if word.isalpha())
tokens = word_tokenize(cleaned_text)

# Remove stopwords and filter out single-letter tokens
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words and len(word) > 1]

# Create bigrams
bi_grams = list(bigrams(filtered_tokens))

# Create a frequency distribution of bigrams
bigram_fd = nltk.FreqDist(bi_grams)

# Create a list of unique words for x and y axes
unique_words = list(set(filtered_tokens))
num_unique_words = len(unique_words)

# Create a matrix to store bigram frequencies
bigram_matrix = np.zeros((num_unique_words, num_unique_words))

# Populate the matrix with bigram frequencies
for bigram, frequency in bigram_fd.items():
    word1, word2 = bigram
    idx1 = unique_words.index(word1)
    idx2 = unique_words.index(word2)
    bigram_matrix[idx1, idx2] = frequency

# Create a heatmap for bigram frequencies
plt.figure(figsize=(10, 10))
plt.imshow(bigram_matrix, cmap='viridis')
plt.colorbar()
plt.xticks(np.arange(num_unique_words), unique_words, rotation=45)
plt.yticks(np.arange(num_unique_words), unique_words)
plt.title('Bigram Frequency Heatmap')
plt.tight_layout()
plt.show()
```

```python
import nltk
from nltk.util import bigrams
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import PyPDF2
import matplotlib.pyplot as plt
import numpy as np
nltk.download('punkt')
nltk.download('stopwords')

# Read the PDF file containing papers
pdf_file_path = r'C:\Users\user\Desktop\499B\BERT.pdf'

# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    text = ''
    with open(pdf_path, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        for page in pdf_reader.pages:
            text += page.extract_text()
    return text

# Extract text from PDF
papers_text = extract_text_from_pdf(pdf_file_path)

# Preprocess and clean the text
cleaned_text = ' '.join(word for word in word_tokenize(papers_text.lower()) if word.isalpha())
tokens = word_tokenize(cleaned_text)

# Remove stopwords and filter out single-letter tokens
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words and len(word) > 1]

# Create bigrams
bi_grams = list(bigrams(filtered_tokens))

# Create a frequency distribution of bigrams
bigram_fd = nltk.FreqDist(bi_grams)

# Select the 100 most frequent bigrams
top_bigrams = bigram_fd.most_common(100)

# Create a list of unique bigrams
unique_bigrams = [bigram for bigram, _ in top_bigrams]

# Create a matrix to store bigram co-occurrence frequencies
num_unique_bigrams = len(unique_bigrams)
bigram_matrix = np.zeros((num_unique_bigrams, num_unique_bigrams))

# Create a mapping of unique bigrams to their indices
bigram_to_idx = {bigram: idx for idx, bigram in enumerate(unique_bigrams)}

# Populate the matrix with bigram co-occurrence frequencies
for token1, token2 in bi_grams:
    if (token1, token2) in bigram_to_idx:
        idx1 = bigram_to_idx[(token1, token2)]
        bigram_matrix[idx1, idx1] += 1

# Create a heatmap for the top 100 bigram co-occurrence frequencies
plt.figure(figsize=(12, 12))
plt.imshow(bigram_matrix, cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.title('Top 100 Bigram Co-occurrence Heatmap')
plt.xticks(range(num_unique_bigrams), unique_bigrams, rotation=90)
plt.yticks(range(num_unique_bigrams), unique_bigrams)
plt.tight_layout()
plt.show()
```
